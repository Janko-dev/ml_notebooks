```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer

df = load_breast_cancer(as_frame=True).frame

df
```

```python
from sklearn.preprocessing import Normalizer

df.iloc[:, :-1] = Normalizer().transform(df.iloc[:, :-1])
df
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

df_pca = PCA(n_components=2).fit_transform(df.iloc[:, :-1])
df_lda = LinearDiscriminantAnalysis(n_components=1).fit_transform(df.iloc[:, :-1], df.iloc[:, -1])

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=df.iloc[:, -1], ax=axes[0])
axes[0].set_title("PCA")
sns.scatterplot(x=np.linspace(0, 1, df_lda.shape[0]), y=df_lda[:, 0], hue=df.iloc[:, -1], ax=axes[1])
axes[1].set_title("LDA")
```

```python
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

df_pca = PCA(n_components=.95).fit_transform(df.iloc[:, :-1])
X_train, X_test, y_train, y_test = train_test_split(df_pca, df.iloc[:, -1], stratify=df.iloc[:, -1])
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
log_reg = LogisticRegression().fit(X_train, y_train)
print(f"{log_reg.score(X_test, y_test)=}")

percep = Perceptron().fit(X_train, y_train)
print(f"{percep.score(X_test, y_test)=}")

svm = SVC().fit(X_train, y_train)
print(f"{svm.score(X_test, y_test)=}")
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

df_lda = LinearDiscriminantAnalysis().fit_transform(df.iloc[:, :-1], df.iloc[:, -1])
X_train, X_test, y_train, y_test = train_test_split(df_lda, df.iloc[:, -1], stratify=df.iloc[:, -1])
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
log_reg = LogisticRegression().fit(X_train, y_train)
print(f"{log_reg.score(X_test, y_test)=}")

percep = Perceptron().fit(X_train, y_train)
print(f"{percep.score(X_test, y_test)=}")

svm = SVC().fit(X_train, y_train)
print(f"{svm.score(X_test, y_test)=}")
```
